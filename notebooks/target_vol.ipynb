```python
contracts = {
    'NQ': {'multiplier': 20,
          'trades_per_day': 4.5,
          'atr': 38, 
          'atr_multiple': 1},
    'ES': {'multiplier': 50,
          'trades_per_day': .8,
           'atr': 12,
           'atr_multiple': 3
          },
    'GC': {'multiplier': 100,
          'trades_per_day': 3,
          'atr': 4.5,
          'atr_multiple': 2},
    'YM': {'multiplier': 5,
          'trades_per_day': 1.5,
          'atr': 135,
          'atr_multiple': 2}
}
multiplier_dict = {2: 1.35, 3:1.5, 4: 1.65}
target_vol = .5
div_multiplier = multiplier_dict[len(contracts)]
alloc = 1/len(contracts)
cash = 104500
```

```python
def number_of_contracts(params: dict, target_vol: float = target_vol, cash: int = cash, alloc: float = alloc):
    daily_vol = target_vol / 16
    daily_cash_alloc = daily_vol * cash * alloc
    cash_alloc_per_trade = daily_cash_alloc / (params['trades_per_day'] ** .5)
    points_alloc_per_trade = cash_alloc_per_trade / params['multiplier']
    contracts = points_alloc_per_trade / ((params['atr'] / 1.5) * params['atr_multiple'])
    return round(contracts * div_multiplier, 1)
```

```python
out = {}
for c, p in contracts.items():
    out[c] = number_of_contracts(p, target_vol=.5)
out
```
